# La Banca Central Bank dataset

```python
import pandas as pd

# La Banca Central Bank dataset

# upload data from csv
fraud_1_df = pd.read_csv('c:/Lab_Files/Data/FraudDatasetSubset_1.csv')
fraud_2_df = pd.read_csv('c:/Lab_Files/Data/FraudDatasetSubset_1.csv')

# merge the two datasets
fraud_merged = fraud_1_df.merge(fraud_2_df, left_on='nameOrig', right_on='nameOrig')

# fill all missing values with 0s
fraud_merged.fillna(0, inplace=True)

# double check to ensure merge was done properly
print(f'original dataset sizes:  fraud_1_df={fraud_1_df.shape},  fraud_2_df={fraud_2_df.shape}' )
print(f'merged dataset size:  fraud_merged={fraud_merged.shape}')

# print the first 10 rows of data
print(fraud_merged.head(10))
```

# Western  Forest Service

```python
import pandas as pd

# Western  Forest Service

# upload data from csv
forest_1_df = pd.read_csv('c:/Lab_Files/Data/ForestCoverTypeDatasetSubset_1.csv')
forest_2_df = pd.read_csv('c:/Lab_Files/Data/ForestCoverTypeDatasetSubset_2.csv')

# merge the two datasets
forest_merged = forest_1_df.merge(forest_2_df, left_on='key', right_on='key')

# fill all missing values with 0s
forest_merged.fillna(0, inplace=True)

# double check to ensure merge was done properly
print(f'original dataset sizes:  forest_1_df={forest_1_df.shape},  forest_2_df={forest_2_df.shape}' )
print(f'merged dataset size:  forest_merged={forest_merged.shape}')

# print the first 10 rows of data
print(forest_merged.head(10))
```
